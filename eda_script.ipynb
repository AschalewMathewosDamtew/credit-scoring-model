```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('data.csv')

# 1. Overview of the Data
print("Dataset Information:")
print(df.info())  # Data types and non-null counts

print("\nFirst 5 rows of the dataset:")
print(df.head())  # Display the first 5 rows

print("\nShape of the dataset:")
print(df.shape)  # Number of rows and columns

# 2. Summary Statistics
print("\nSummary Statistics for Numerical Features:")
print(df.describe())  # Summary statistics for numerical features

# 3. Distribution of Numerical Features
print("\nDistribution of 'Amount' and 'Value':")
plt.figure(figsize=(14, 6))

# Plot for Amount
plt.subplot(1, 2, 1)
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Distribution of Amount')

# Plot for Value
plt.subplot(1, 2, 2)
sns.histplot(df['Value'], bins=50, kde=True)
plt.title('Distribution of Value')

plt.tight_layout()
plt.show()

# 4. Distribution of Categorical Features
print("\nCategorical Feature Distribution:")
plt.figure(figsize=(14, 6))

# Plot for ProductCategory
plt.subplot(1, 2, 1)
df['ProductCategory'].value_counts().plot(kind='bar')
plt.title('Product Category Distribution')

# Plot for ChannelId
plt.subplot(1, 2, 2)
df['ChannelId'].value_counts().plot(kind='bar')
plt.title('Channel Id Distribution')

plt.tight_layout()
plt.show()
```

```python
# 5. Correlation Analysis (Handling non-numeric columns)
# Drop CountryCode column as it provides no useful information
# Selecting only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Now compute the correlation matrix
print("\nCorrelation Matrix:")
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# 6. Identifying Missing Values
print("\nMissing Values in each column:")
print(df.isnull().sum())

# 7. Outlier Detection using Boxplots
plt.figure(figsize=(12, 6))

# Boxplot for Amount
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Amount'])
plt.title('Outliers in Amount')

# Boxplot for Value
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Value'])
plt.title('Outliers in Value')

plt.tight_layout()
plt.show()
```
